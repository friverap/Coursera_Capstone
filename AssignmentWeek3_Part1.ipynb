# Sementing and Clustering Neigborhoods in the city of Toronto (Part 1): Scraping data from Wikipedia page

```python
#import libreries
import pandas as pd
import numpy as np
import requests
```

```python
#Get the HTML page of Wikipedia
postal_code_CA = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
CA = requests.get(postal_code_CA)
CA_raw = pd.read_html(CA.text)
CA_raw
```

```python
#Create a dataframe
df = pd.DataFrame(CA_raw[0])
df.head(12)
```

```python
#rewrite the columns name
rw = {'Postal Code': 'PostalCode', 'Borough':'Borough', 'Neighbourhood': 'Neighborhood'}
df = df.rename(columns=rw)
df.head(12)
```

```python
#Ignore cells with a borough that is Not assigned
df = df[df['Borough'] != 'Not assigned']
df.head(12)
```

```python
#Check if there are a duplicate PostalCode
if True not in df.duplicated(['PostalCode']).tolist():
    print('No duplicates in the Postal Code column.')
else: 
    print('You should combine rows with duplicates in the Postal Code column')
```

```python
#check if there are rows that have a borough but not an assigned neighborhood
if df[(df['Neighborhood'].isna())&(df['Borough'] != 'Not assigned')].shape[0] == 0:
    print('There are no rows that have a borough but not an assigned neighborhood.')
```

```python
#reset de index of the data frame
ca_df = df.reset_index(drop=True)
ca_df.head(12)
```

```python
#print the number of rows of the new dataframe
print('The number of rows in the final dataframe:', ca_df.shape[0])
```

```python
#save the final data frame
ca_df.to_csv('CapstoneWeek3.csv')
```
